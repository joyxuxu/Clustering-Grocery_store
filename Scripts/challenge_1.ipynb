# Finding the Total No. of items purchased by each customer

```python
import pandas as pd
purchase_history_df = pd.read_csv('../Data/prepared_purchased_history.csv')

col_list = purchase_history_df.columns.tolist()
col_list.remove('id')
purchase_history_df['purchased_item_count'] = purchase_history_df[col_list].sum(axis=1)
purchase_history_df
```

# Finding the customer_id  with max. purchasal count

```python
max_customer_index = purchase_history_df['purchased_item_count'].idxmax()
Max_purchasal_cust_id = purchase_history_df['id'].iloc[max_customer_index]
print ("The Customer id who bought the most items overall in her lifetime is: {}".format(Max_purchasal_cust_id))
```

# Finding the list of customers who bought each item max. No. of times

```python
output_df = pd.DataFrame(columns=['Item_id', 'customer_id'])
for col in col_list:
    each_max_cust_index = purchase_history_df[col].idxmax()
    each_item_max_cust_id = purchase_history_df['id'].iloc[each_max_cust_index]
    row = pd.Series([int(col.split('_')[-1]), each_item_max_cust_id],index=['Item_id', 'customer_id'])
    output_df = output_df.append(row,ignore_index=True)
    output_df['Item_id'] = pd.to_numeric(output_df['Item_id'])

items_df = pd.read_csv('../Data/item_to_id.csv')
final_output_df = pd.merge(output_df, items_df, on='Item_id')
final_output_df
```

### Saving the results to a csv file 

```python
final_output_df.to_csv('../Data/each_item_max_purchase_customer_id.csv',index=False)
```
